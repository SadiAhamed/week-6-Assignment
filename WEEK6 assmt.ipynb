```python
import pandas as pd
import numpy as np
import statistics as stats
data=pd.read_csv(r"D:\activity\StudentsPerformance.csv")
df=pd.DataFrame(data)
data5=df.groupby("gender").describe()
data5
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', figsize=(10,10))
plt.xticks(rotation=0)
plt.title('Gender vs Average score',c='c',fontsize=15)
plt.xlabel('Gender',c='c',fontsize=13)
plt.ylabel('Average',c='c',fontsize=13)
plt.grid()
plt.show()
```

## RESULT
#### 1.We can see that females have more average score in reading and writing than that of males.
#### 2.In the case of math average score males have more average score than females.
#### 3.By taking the over all average of all the 3 subjects, females have more average score than males
